__Handling Missing Values in NumPy__

 Why Missing Values Matter:
 
Missing data is common in real-world datasets (due to sensor failure, manual entry errors, etc.). NumPy doesn't have a built-in missing value type like NaN in pandas, but it can work with np.nan (Not a Number) to represent missing values in float arrays.



```python
import numpy as np

#  Representing Missing Values np.nan
a=np.array([1,2,3,4,np.nan,6,7,np.nan])
a
```

```python
# Detecting Missing Values np.isnan(array)
# it gives the Boolean Array with True at Nan Values
Missing=np.isnan(a)
print("Missing values ",Missing)
```

```python
# Filtering Valid  Values
# Neglecting those nan Values ~(np.isnan(a))
valid=a[~(np.isnan(a))]
print(valid)
```

```python
#  Common Operations with Missing Values
mean=a.mean()
print(mean)
# we can observe that calculating the mean 
# with nan values : we always get nan 
# which is most important to check
```

```python
# so to Ignore this we use nan versions of operations
print(np.nanmean(a) )  # Mean ignoring NaN
print(np.nanstd(a)  )  # Standard deviation ignoring NaN
print(np.nansum(a)  )  # Sum ignoring NaN
print(np.nanprod(a))   # Product Ignoring nan
print(np.nanmedian(a))   # medianIgnoring nan
```

```python
arr = np.array([7, np.nan, 2, 9, np.nan, 4])

"""# 1. Print how many missing values are there
# 2. Replace missing values with the median
# 3. Compute the sum ignoring NaNs
"""

#1 
print(np.isnan(arr))
print(arr[np.isnan(arr)])

#2 
arr2=arr.copy()
arr2[np.isnan(arr2)]=np.nanmedian(arr2)
print(arr2)

#3
print(np.nansum(arr))
```

```python
"""Replace np.nan with Where()

Using np.where( valuesToReplace , single_value to Replace , Array)

return the new array with changed values
""" 

data = np.array([10.0, 20.0, np.nan, 40.0, np.nan])
filled = np.where(np.isnan(data),0,data)
print(filled)
```

```python
"""
Replace np.nan with np.nan_to_num()

Using np.nan_to_num( array, single_value_toreplace)

return the new array with changed values
""" 
data = np.array([10.0, 20.0, np.nan, 40.0, np.nan])
cleaned = np.nan_to_num(data, nan=0)
print(cleaned)
```

```python
# Plotting x=y with Numpy 
import matplotlib.pyplot as p
import numpy as np

x=np.linspace(-10,10,100)
y = x
# same should be same of array x and y
p.plot(x,y)
```

```python
 # Plotting y=sinx with Numpy 
import matplotlib.pyplot as p
import numpy as np

x=np.linspace(-10,10,100)
y = np.sin(x)

# same should be same of array x and y
p.plot(x,y)
p.title("Sine Wave")
p.xlabel("x")
p.ylabel("sin(x)")
p.grid(True)
p.show()
```

```python
# Sigmoid Function Graph 
# Sigmoid Function With Numpy:
import matplotlib.pyplot as plt
import numpy as np
def Sigmoid(numpy_array):
    Solution=1/(1+np.exp(-(numpy_array)))
    return Solution

#1 Example
x=np.linspace(-10,10,100)
y=Sigmoid(x)

plt.plot(x,y,label='sigmoid',color='Black')
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y=sigmoid")
plt.show()
```
