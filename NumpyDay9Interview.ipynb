# Meshgrids

```python
"""Theroy :
np.meshgrid() takes two or more 1D arrays and returns coordinate matrices from them. 
Its commonly used to create a grid of points over a 2D space.
it  creates 2D grid arrays from your 1D arrays.


Used For :
1. 3D plotting (matplotlib)

2. Evaluating functions over a grid (e.g., Z = f(X, Y))

3. Contour plots, surface plots, vector fields"""
```

```python
a=np.linspace(0,2,3)
b=np.linspace(0,2,3)
print(a,b)
# Creating the grid
# generally we store it as  xx and yy
#This creates 2D grid arrays from your 1D a and b.
xx,yy=np.meshgrid(a,b)
print(xx) 
print(yy) 
# it returns two array which creates grid together
"""
Explanation of xx and yy:
xx repeats the values from a in each row → this controls the X-coordinates

yy repeats the values from b in each column → this controls the Y-coordinates

Together:

At any position (i, j), the point is (xx[i][j], yy[i][j])

So you're building a grid of (x, y) pairs like:

(0,0) (1,0) (2,0)
(0,1) (1,1) (2,1)
(0,2) (1,2) (2,2)

"""
```

```python
# to view points of grid
"""
(0,0) (1,0) (2,0)
(0,1) (1,1) (2,1)
(0,2) (1,2) (2,2)
"""
import matplotlib.pyplot as plt
plt.scatter(xx,yy)
```

```python
# now
#  We X (xx) and Y (yy) as 2D grids —
#  so the third dimension Z can be easily created
#  using any formula that uses X and Y.


# let us suppose z= x^2 +y^2 
# TO make formula Z=xx**2 +yy**2 (because i take xx and yy as my name of 2d arrays )

Z=xx**2 +yy**2
print(Z)
```

# Q. What are the structured Numpy array?

structured NumPy arrays (also called record arrays) let you store heterogeneous data types (like columns in a table) within a single NumPy array.

It’s an array where each element has named fields, and each field can have its own data type (like a table row with columns).

```python
# We have Understand that , we can create the Homogenous 
# Numpy Array . like
import numpy as np
x=np.array([1,2,3,4,5])
x.dtype

# It has only one data type Elements 
```

```python
# what if we Try TO create the Numpy Array of 
# Different Data type ELements
import numpy as np
x=np.array([1,"hello",3.1837,True,"String"])
print(x)
print(x.dtype)

# It Still Create the Array of Datatype U32
# if we check Every Element type
# it is numpy.str it means numpy automatically convert all 
# Element in String even we provide the Different Datatype Elements
# so logically it is a Homogenous array of String Datatype

print(type(x[0]))
print(type(x[1]))
print(type(x[2]))
print(type(x[3]))
```

```python
# What we can Do To Create DIfferent Datatype Array?
# ===> how to Create Heterogenous Array
# we USe Structured Array...

"""
We define the  Structure with np.dtype() function
Each entry is a tuple:

('field_name', 'data_type')

"""

dt=np.dtype(
    [
        ('name',"U32"),
        ('age',np.int32),
        ('weight',np.float64)
    ]
)
```

```python
# this Is My Structure dtype()
dt
```

```python
# now we create the Structured array
"""
Now use np.array(..., dtype=dt) to build the actual data:"""

array=np.array([
    ('varun',18,55.7),
    ('prince',17,50)

],dtype=dt)

print(array)
array.dtype
```

```python
# to access Elements from the Structured Array
print(array[0]) # first Element (1 row)
print(array[1]) # Second Element (2 row)
```

```python
# TO get the COlumnn we can use the "FieldName"
print(array["name"])
print(array["age"])
print(array["weight"])
```

# Save and Load Using 

- .npy (Single Object)

Saving and loading NumPy objects is super important for working with large datasets, models, or intermediate results — it saves you from recalculating things every time.

```python
"Use this when you want to save one NumPy array."
# SAVE
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr)
np.save('NumpyDay9Interviewq.npy', arr)
# After saving this we can load it anywhere ,
# like in another file or notebook 
# where we have need of it,
```

```python
# LOAD
LoadedArray=np.load("NumpyDay9Interviewq.npy")
print(LoadedArray)
```

- .npz (Multiple Numpy Object)

```python
# save
a = np.arange(5)
b = np.linspace(0, 1, 5)

np.savez('NumpyDay9Interviewq.npz', array1=a, array2=b)
```

```python
# Load
data = np.load('NumpyDay9Interviewq.npz')

print(data['array1'])  # Access by name
print(data['array2'])
```
