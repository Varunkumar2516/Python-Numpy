### `Q-11:` Softmax function

Create a Python function to calculate the Softmax of the given numpy 1D array. The function only accepts the numpy 1D array, otherwise raise error.

$$\Large \sigma(\vec z)_i = \frac{e^{z_i}}{\sum_{j=i}^{K}{e^{z_j}}}$$

https://en.wikipedia.org/wiki/Softmax_function



**Example 1:**

Input:
```bash
[86.03331084 37.7285648  48.64908087 87.16563062 38.40852563 37.20006318]
```

Output:
```bash
[2.43733249e-01, 2.56112115e-22, 1.41628284e-17, 7.56266751e-01,
       5.05514197e-22, 1.50974911e-22]
```

**Example 2:**

Input:
```bash
[33.17344305 45.61961654 82.05405781 80.9647098  68.82830233 91.52064278]
```

Output:
```bash
[4.57181035e-26, 1.16249923e-20, 7.73872596e-05, 2.60358426e-05,
       1.39571531e-10, 9.99896577e-01]
```

````python
"""Theroy :
The softmax function is used in machine learning and deep learning, especially in classification problems (like image or text classification).

It turns a list of raw numbers (called logits or scores) into probabilities — numbers between 0 and 1 that add up to 1.



"""
````

````python
#Code here
def Softmax(numpy_array):
    # there may Overflow if number inside the Numpy array array is very large like >10000 so exp(10000)-> infinite so to overcome this situation
    # we can substract the maximum number from whole array so that we get the smalledformat array and then apply the expo to it
    smallerformat=numpy_array - np.max(numpy_array)
    Exponent=np.exp(smallerformat)
    sum=np.sum(Exponent)
    Solution=Exponent /sum
    return Solution

NumpyArray=np.array( [86.03331084 ,37.7285648 ,48.64908087 ,87.16563062 ,38.40852563 ,37.20006318])
print(Softmax(NumpyArray))
````

### `Q-13:` Dates

Create a python function named **date_array** that accepts two dates as string format and returns a numpy array of dates between those 2 dates. The function only accept 2 strings, otherwise raise error. The date format should be like this only: `2022-12-6`. The end date should be included and for simplicity, choose dates from a same year.



**Example 1:**

Input:
```bash
date_array(start = '2020-09-15', end = '2020-09-25')
```

Output:
```bash
['2020-09-15', '2020-09-16', '2020-09-17', '2020-09-18',
 '2020-09-19', '2020-09-20', '2020-09-21', '2020-09-22',
 '2020-09-23', '2020-09-24', '2020-09-25']
```

**Example 2:**

Input:
```bash
date_array(start = '2022-12-01', end = '2022-12-06')
```

Output:
```bash
['2022-12-01', '2022-12-02', '2022-12-03', '2022-12-04', '2022-12-05', '2022-12-06']
```

**Example 3:**

Input:
```bash
date_array(start = '2020-11-25', end = '2020-11-30')
```

Output:
```bash
['2020-11-25', '2020-11-26', '2020-11-27', '2020-11-28',
 '2020-11-29', '2020-11-30']
```

````python
#Code here
import datetime
def date_array(start,end):
    if not isinstance(start,str) or not isinstance(end,str):
        raise ValueError("Enter the Date Correctly")
    # date is in format DD-MM-YY
    format="%Y-%m-%d"
    # creating the date object from the string 
    obj1=datetime.datetime.strptime(start,format)
    obj2=datetime.datetime.strptime(end,format)
    difference=obj2-obj1
    Days=difference.days+1 # for end date we add +1 to it 
    Date_list=[(obj1 + datetime.timedelta(days=i)).strftime(format) for i in range(Days) ]
    return Date_list

# Example 1
Dates=date_array(start = '2020-09-15', end = '2020-09-25')
print(Dates)
````

````python
# Example 2:
Dates=date_array(start = '2022-12-01', end = '2022-12-06')
print(Dates)
````

````python
# Example 3:
Dates=date_array(start = '2020-11-25', end = '2020-11-30')
print(Dates)
````

###`Q-17:` Given two arrays of same shape make an array of max out of two arrays. (Numpy way)
```
a=np.array([6,3,1,5,8])
b=np.array([3,2,1,7,2])

Result-> [6 3 1 7 8]
```

````python
#Code here
"""
np.maximum(array1, array2) compares element by element.

Returns a new array where each element is the maximum of the two at that position.
"""
def Max(Array1,Array2):
    if Array1.shape !=Array2.shape:
        raise ValueError("Enter ")
    return np.maximum(Array1,Array2)
a=np.array([6,3,1,5,8])
b=np.array([3,2,1,7,2])
max=Max(a,b)
print(max)
````

https://en.wikipedia.org/wiki/Normalization_(statistics)

There are different form of normalisation for this question use below formula.

$$\large X_{normalized} = \frac{X - X_{min}}{X_{max} - X_{min}}$$

```python
arr1=np.random.randint(low=1, high=10000, size=40).reshape(8,5)
```


````python
# Theory :
"""Normalization transforms data to a common scale (often between 0 and 1).
 This makes comparison easier and removes units or differences in scales.

 Explanation of the terms:
X: the original array (or an individual element of it)

X_min: the minimum value in the array

X_max: the maximum value in the array

X_normalized: the normalized result, which lies between 0 and 1
"""
````

````python
# Given
arr1=np.random.randint(low=1, high=10000, size=40).reshape(8,5)
arr1
````

````python
#Code here
#1 :Fetch Every Alternate Column of Array
def Alternate(array):
    # Alternate
    print(array[:,::2])

Alternate(arr1)
````

````python
#2 :Normalize the Given Array

def Normalize(array):
    Min=array.min()
    Max=array.max()
    if Min==Max:
        return ValueError("Max != Min (max Cannot be same as min) ")
    Normalize=(array-Min)/(Max-Min)
    return Normalize

print(Normalize(arr1))
````

###`Q-19:` Write a function which will accept 2 arguments.
First: A 1D numpy array arr

Second: An integer n {Please make sure n<=len(arr)}

Output: The output should be the nth largest item out of the array
```
# Example1 : arr=(12,34,40,7,1,0) and n=3, the output should be 12
# Example2 : arr=(12,34,40,7,1,0) and n=1, the output should be 40
```

````python
#Code here
"""
We can remove a single element from a NumPy array

np.delete(arr, index) 

arr: the original array
index: the index of the element to remove"""
# it does not change the original array ,it gives the copy of the array
# so to change assign it back to original array
"""
To Get the maximum element Index IN Numpy Element
np.argmax(arr) → gives the index of the first occurrence of the maximum value.
"""
def NthLargest(array,n):

    if n==1:
       return np.max(array)
    else:
        for i in range(1,n):
          array=np.delete(array,np.argmax(array))
        return np.max(array)
  
````

````python
#testing
arr=np.array([12,34,40,7,1,0]) 
print(NthLargest(arr,3))

#testing
arr=np.array([12,34,40,7,1,0]) 
print(NthLargest(arr,1))

#testing
arr=np.array([12,34,40,7,1,0]) 
print(NthLargest(arr,5))
````

###`Q-20:` Create the following pattern without hardcoding. Use only numpy functions and the below input array a.
```
# Input: a = np.array([1,2,3])
# Output: array([1, 1, 1, 2, 2, 2, 3, 3, 3, 1, 2, 3, 1, 2, 3, 1, 2, 3])
```

````python
"""NumPy  has np.append() —That can add the ELements in numpy even if you already created it 

Syntax np.append(Array_name,element_to_Append)"""
````

````python
#code here
def Encoding(array):
    size=array.size
    encoded=np.array([])
    for j in range(size):
        for _ in range(size):
            encoded=np.append(encoded,array[j])
    
    for _ in range(size):
        encoded=np.append(encoded,array)
    return encoded
    


a = np.array([1,2,3])
a.size
Encoded_Array=Encoding(a)
print(Encoded_Array)
print(type(Encoded_Array))
````

````python
# Theory for more better way :
"""
np.repeat(array, 3)	Repeats each element of array three times
np.tile(array, 3)	Repeats the whole array array three times
np.concatenate([A1,A2])	Joins both arrays into one
"""

def Encoding(array):
    Part1=np.repeat(array,3)

    part2=np.tile(array,2)

    Result=np.concatenate([Part1,part2])
    

    return Result


a = np.array([1,2,3])
a.size
Encoded_Array=Encoding(a)
print(Encoded_Array)
print(type(Encoded_Array))
````
