__Graphing with NumPy + Matplotlib__

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Basic Line Plot
x = np.linspace(0, 10, 100)      # 100 points from 0 to 10
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```

```python
#Plotting Multiple Lines
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)', color='blue')
plt.plot(x, y2, label='cos(x)', color='Black')
# plt.legend()
plt.title("Sine and Cosine")
plt.grid(True)
plt.show()
```

```python
# Scatter Plot
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, color='green')
plt.title("Random Scatter")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# Histogram
data = np.random.randn(1000)  # Normal distribution

plt.hist(data, color='black', edgecolor='white')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#Bar Graph
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]

plt.bar(categories, values, color='black')
plt.title("Bar Chart Example")
plt.ylabel("Scores")
plt.show()
```

```python
# Fill Area Under Curve
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.fill_between(x, y, color="Black")
plt.title("Filled Area Under Sine Curve")
plt.show()
```
